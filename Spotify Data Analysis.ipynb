```python
tracks_list = []
for row in data['items']:
    date = row['added_at']
    album_name = row['track']['album']['name']
    album_url = row['track']['album']['external_urls']['spotify']
    total_tracks = row['track']['album']['total_tracks']
    album_id = row['track']['album']['id']
    
    tracks_element = {'date':date,'album_name':album_name,'album_url':album_url,'total_tracks':total_tracks,'album_id':album_id}
    tracks_list.append(tracks_element) 
```

```python
tracks_list
```

```python
album_df = pd.DataFrame(tracks_list)
```

```python
album_df
```

# Artists Data

```python
artist_list = []

for album in data['items']:
    release_date = album['track']['album']['release_date'] 
    album_name = album['track']['album']['name']
    song_name = album['track']['name']
    for artists_list in album['track']['artists']:
        artists_name = artists_list['name']
        artists_id = artists_list['id']
        artists_dict = {'release_date' : release_date,
                        'artists_name':artists_name,
                        'artists_id': artists_id,
                        'song':song_name,
                        'album_name': album_name }
        
        artist_list.append(artists_dict)

```

```python
artist_list
```

```python
artists_data = pd.DataFrame(artist_list) 
```

```python
artists_data
```

```python
song_list = []

for album in data['items']: 
    song_name = album['track']['name']  
    popularity = album['track']['popularity'] 
    song_id = album['track']['id'] 
    song_duration_ms = album['track']['duration_ms'] 
    songs_urls = album['track']['external_urls']['spotify']
#     artists_id = album['track']['album']['artists'][0]['id'] 
    album_id = album['track']['album']['id']
    
    song_dict = {'song_name' : song_name,
                'popularity' : popularity,
                'song_id': song_id,
                'song_duration_ms' : song_duration_ms,
                'song_urls' : songs_urls,
                'artists_id' : artists_id,
                'album_id' : album_id} 
    song_list.append(song_dict) 
```
